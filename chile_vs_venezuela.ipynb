### Chile vs. Venezuela: A Tale of Two South American Economies Under Dictatorship

In 1973, A military coup, led by Augusto Pinochet, took control over Chile, which was experiencing massive inflation under the leadership of Salvadore Allende. Incredibly, the coup leaders decided to listen to economists. Income has seen massive improvedments in Income Per Person since the economists advice was listened to. This is known as "The Miracle of Chile".

Venezuela is the polar opposite. Hugo Chavez became president in 1999 and would eventually becomee a dictator himself. Unlike Chile, he went in a direction opposed to all mainstream economic thought. His successor, Nicolas Maduro has fared no better.

The interesting thing to note is that the official Gini Coefficient for both countries, a measure of inequality, is roughly the same in both countries. However, the size of the economic pie is much larger in Chile than in Venezuela. This has resulted in a massive migration from Venezuela to other countries. In fact, many Venezuelans fled to, and now live in, Chile.

```
#Library Import
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```
#Data Import
income = pd.read_csv('./data/income_per_person_gdppercapita_ppp_inflation_adjusted.csv', index_col='country')
gini = pd.read_csv('./data/gini.csv', index_col='country')
```

```
#Set Row Vars
chileIncome = income.loc['Chile']
venezuelaIncome = income.loc['Venezuela']
chileGini = gini.loc['Chile']
venezuelaGini = gini.loc['Venezuela']
```

```
#Clean Data
chileIncomeFinal = chileIncome['1960':'2019']
venezuelaIncomeFinal = venezuelaIncome['1960':'2019']
chileGiniFinal = chileGini['1960':'2019']
venezuelaGiniFinal = venezuelaGini['1960':'2019']
```

```
#Plot Gini Coefficient
fig= plt.figure(figsize=(10,5))
plt.title('Venezuela Vs. Chile: Gini Coefficient (1960 - 2019)')
plt.style.use('seaborn-darkgrid')
plt.xlabel("Year")
plt.ylabel("Income")
venezuelaGiniFinal.plot(label='Venezuela', color='red', linewidth=1, alpha=0.9)
chileGiniFinal.plot(label='Chile', color='blue', linewidth=1, alpha=0.9)
plt.legend()
```

```
#Plot Income
fig= plt.figure(figsize=(10,5))
plt.title('Venezuela Vs. Chile: Income Per Person, PPP, Inflation Adjusted (1960 - 2019)')
plt.style.use('seaborn-darkgrid')
plt.xlabel("Year")
plt.ylabel("Income")
venezuelaIncomeFinal.plot(label='Venezuela', color='red', linewidth=1, alpha=0.9)
chileIncomeFinal.plot(label='Chile', color='blue', linewidth=1, alpha=0.9)
plt.legend()
```
